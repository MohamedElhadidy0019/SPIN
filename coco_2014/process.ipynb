```python
import math

def calculate_fy(dfov, width, height):
    """
    Calculate fy (vertical focal length) from diagonal FOV, width, and height.

    Parameters:
    dfov (float): Diagonal field of view in degrees.
    width (int): Image width in pixels.
    height (int): Image height in pixels.

    Returns:
    float: fy (focal length in the vertical direction).
    """
    # Convert DFOV to radians
    dfov_rad = math.radians(dfov)
    
    # Calculate aspect ratio
    aspect_ratio = width / height
    
    # Calculate vertical FOV (v) from DFOV
    v = 2 * math.atan(math.tan(dfov_rad / 2) / math.sqrt(1 + aspect_ratio**2))
    
    # Calculate fy using the given formula
    fy = height / (2 * math.tan(v / 2))
    return fy

# Inputs
dfov = 160  # Diagonal field of view in degrees
width = 640  # Width of the image in pixels
height = 480  # Height of the image in pixels

# Calculate fy
fy = calculate_fy(dfov, width, height)
print(f"Calculated fy: {fy}")
```
